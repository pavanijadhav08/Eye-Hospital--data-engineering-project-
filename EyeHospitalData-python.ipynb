```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake = Faker()

emr_data = []

for i in range(500):
    emr_data.append({
        "Patient_ID": i+1,
        "Name": fake.name(),
        "Age": random.randint(1,80),
        "Gender": random.choice(["Male","Female"]),
        "Phone": fake.phone_number(),
        "Disease": random.choice(["Cataract","Glaucoma","Diabetes","Normal","Retina"]),
        "Visit_Date": fake.date(),
        "Doctor": random.choice(["Dr.A","Dr.B","Dr.C"]),
        "Hospital": "Eye Hospital"
    })

emr_df = pd.DataFrame(emr_data)

emr_df.to_excel("emr_data.xlsx", index=False)

print("EMR Data Generated")
emr_df.head()
```

```python
#Data Collection (Load Excel Files)
emr = pd.read_excel("emr_data.xlsx")
camp = pd.read_excel("camp_data.xlsx")

print("EMR Data Shape:", emr.shape)
print("Camp Data Shape:", camp.shape)

emr.head()
```

```python
#Data Profiling
print("----- EMR INFO -----")
print(emr.info())

print("\n----- EMR STATISTICS -----")
print(emr.describe())

print("\n----- MISSING VALUES -----")
print(emr.isnull().sum())
```

```python
#camp data
print(camp.info())
print(camp.describe())
print(camp.isnull().sum())
```

```python
# Handle missing values

emr["Age"] = emr["Age"].fillna(emr["Age"].mean())

emr["Name"] = emr["Name"].fillna("Unknown")

emr["Disease"] = emr["Disease"].fillna("Not Specified")

# Remove duplicate rows
emr = emr.drop_duplicates()

print("Data Cleaning Completed")

emr.head()
```

```python
#Schema Design
schema = {
"Patient_ID":"Integer",
"Name":"String",
"Age":"Integer",
"Gender":"String",
"Phone":"String",
"Disease":"String",
"Visit_Date":"Date",
"Doctor":"String",
"Hospital":"String"
}

print("Schema Design")
print(schema)
```

```python
#Data Mapping (Camp → Hospital)
mapping = {
"Name":"Name",
"Age":"Age",
"Gender":"Gender",
"Camp_Date":"Visit_Date",
"Doctor":"Doctor"
}

print("Data Mapping")
print(mapping)
```

```python
#ETL Pipeline
def extract():
    emr = pd.read_excel("emr_data.xlsx")
    camp = pd.read_excel("camp_data.xlsx")
    return emr, camp
```

```python
# Transform
def transform(emr, camp):

    emr = emr.drop_duplicates()
    camp = camp.drop_duplicates()

    emr["Age"] = emr["Age"].fillna(emr["Age"].mean())
    camp["Age"] = camp["Age"].fillna(camp["Age"].mean())

    return emr, camp

```

```python
def load(emr,camp):

    emr.to_csv("clean_emr_data.csv",index=False)

    camp.to_csv("clean_camp_data.csv",index=False)

    print("ETL Process Completed Successfully")
```

```python
#Run ETL Pipeline
emr,camp = extract()

emr_clean,camp_clean = transform(emr,camp)

load(emr_clean,camp_clean)
```
